```python
# Strength Test
# This was never used but my ego tells me to keep it
def findTheBiggest(name):
    Bug = 0
    Dark = 0
    Dragon = 0
    Electric = 0
    Fairy = 0
    Fighting = 0
    Fire = 0
    Flying = 0
    Ghost = 0
    Grass = 0
    Ground = 0
    Ice = 0
    Normal = 0
    Poison = 0
    Psychic = 0
    Rock = 0
    Steel = 0
    Water = 0
    
    for i in range(len(bug)):
        Bug += bug.at[i, name]
        newDF = pd.DataFrame([Bug], columns = ["Bug"])
        
    for i in range(len(dark)):
        Dark += dark.at[i, name]
        newDF["Dark"] = Dark
        
    for i in range(len(dragon)):
        Dragon += dragon.at[i, name]
        newDF["Dragon"] = Dragon
        
    for i in range(len(electric)):
        Electric += electric.at[i, name]
        newDF["Electric"] = Electric
    
    for i in range(len(fairy)):
        Fairy += fairy.at[i, name]
        newDF['Fairy'] = Fairy
    
    for i in range(len(fighting)):
        Fighting += fighting.at[i, name]
        newDF['Fighting'] = Fighting
    
    for i in range(len(fire)):
        Fire += fire.at[i, name]
        newDF['Fire'] = Fire
    
    for i in range(len(flying)):
        Flying += flying.at[i, name]
        newDF['Flying'] = Flying
    
    for i in range(len(ghost)):
        Ghost += ghost.at[i, name]
        newDF['Ghost'] = Ghost
    
    for i in range(len(grass)):
        Grass += grass.at[i, name]
        newDF['Grass'] = Grass
    
    for i in range(len(ground)):
        Ground += ground.at[i, name]
        newDF['Ground'] = Ground
    
    for i in range(len(ice)):
        Ice += ice.at[i, name]
        newDF['Ice'] = Ice
    
    for i in range(len(normal)):
        Normal += normal.at[i, name]
        newDF['Normal'] = Normal
    
    for i in range(len(poison)):
        Poison += poison.at[i, name]
        newDF['Poison'] = Poison
    
    for i in range(len(psychic)):
        Psychic += psychic.at[i, name]
        newDF['Psychic'] = Psychic
    
    for i in range(len(rock)):
        Rock += rock.at[i, name]
        newDF['Rock'] = Rock
    
    for i in range(len(steel)):
        Steel += steel.at[i, name]
        newDF['Steel'] = Steel
    
    for i in range(len(water)):
        Water += water.at[i, name]
        newDF['Water'] = Water

    Bug /= len(bug)
    Dark /= len(dark)
    Dragon /= len(dragon)
    Electric /= len(electric)
    Fairy /= len(fairy)
    Fighting /= len(fighting)
    Fire /= len(fighting)
    Flying /= len(flying)
    Ghost /= len(ghost)
    Grass /= len(grass)
    Ground /= len(ground)
    Ice /= len(ice)
    Normal /= len(normal)
    Poison /= len(poison)
    Psychic /= len(psychic)
    Rock /= len(rock)
    Steel /= len(steel)
    Water /= len(water)
    newDF = pd.DataFrame([Bug, Dark, Dragon, Electric, Fairy, Fighting, Fire, Flying, Ghost, Grass, Ground,
           Ice, Normal, Poison, Psychic, Rock, Steel, Water], columns = [name], index = ['Bug', 'Dark', 'Dragon', 'Electric', 'Fairy', 'Fighting', 'Fire',
       'Flying', 'Ghost', 'Grass', 'Ground', 'Ice', 'Normal', 'Poison',
       'Psychic', 'Rock', 'Steel', 'Water'])
    return(newDF)
```

```python

# KNearest Neighbors algorithmn for predicting type 1
X = pokemon_data[["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
Y = pokemon_data["Type 1"]

xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.2)

for i in range(1,50,2):
    classifier = KNeighborsClassifier(n_neighbors = i)
    
    classifier.fit(xTrain, yTrain)
    predictions = classifier.predict(xTest)

    accuracy = accuracy_score(yTest, predictions)
    print(i)
    print(accuracy * 100)
    print()
    # 7-9 is the best amount of neighbors
    
classifier = KNeighborsClassifier(n_neighbors = 8)
classifier.fit(xTrain, yTrain)
predictions = classifier.predict(xTest)

accuracy = accuracy_score(yTest, predictions)
print(accuracy * 100)# 24.375% accuracy
```

```python
#Confusion matrix for the KNearest Neighbors algorithm
ConfusionMatrixDisplay.from_estimator(classifier, xTest, yTest)
plt.xticks(rotation = 55)
#plt.savefig("MachineLearning.png")
```

```python
# ANOVA Testing
fireMask1 = fire.mask(fire.Generation != 1)
group1 = fireMask1.Attack
group1 = group1.dropna()

fireMask2 = fire.mask(fire.Generation != 2)
group2 = fireMask2.Attack
group2 = group2.dropna()

fireMask3 = fire.mask(fire.Generation != 3)
group3 = fireMask3.Attack
group3 = group3.dropna()


# Performing one-way ANOVA
f_statistic, p_value = scipy.stats.f_oneway(group1, group2, group3)


# F-Statistic shows how different the groups are from one another, a higher F-Statistic means more different goes from 0-infinity
print("F-statistic:", f_statistic)
# P-Value will tell us if we can reject the null hypothesis if it is below 0.05 which it is not so we accept that the groups are similar
print("p-value:", p_value)
```
